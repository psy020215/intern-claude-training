### **# Question 1-2** <a name="issue-1_2"></a>
Prompting Claude Code is a skill. Three principles to build that skill:

**Principle 1: Be specific about the task.** Vague prompts produce vague code. Say exactly what input the function takes, what it should return, and any edge cases that matter. If you want a script that reads a GFF3 file and counts features by type, say that — not just "write a GFF parser."

**Principle 2: Give context about your files and environment.** Claude Code can see your files, but it works faster when you point it in the right direction. Mention which file to look at, what format the data is in, or what library you're already using.

**Principle 3: Iterate — your first prompt is a draft.** Don't expect a perfect solution on the first try. Start with a clear first prompt to get a working skeleton, then follow up with refinements: *"That works, but can you also handle malformed lines without crashing?"*
