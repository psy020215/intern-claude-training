## 1. What is Claude Code?

Claude Code is a **command-line tool** — not a chatbot you open in a browser. It runs in your terminal alongside your code.

Key differences from a chatbot:

| Chatbot | Claude Code |
|---|---|
| You paste code in manually | It reads your files directly |
| It can only suggest code | It can write, run, and edit files |
| No project context | It sees your full directory structure |
| New session = blank slate | It holds context across your session |

### Starting Claude Code

Open your terminal in a project directory and type:

```bash
claude
```

That's it. Claude Code starts, reads your project structure, and waits for your first prompt. From that point, you're having a conversation with an agent that has full access to your files.

## 2. How to Prompt Effectively

Prompting Claude Code is a skill. Three principles to build that skill:

### Principle 1: Be specific about the task
Vague prompts produce vague code. Say exactly what input the function takes, what it should return, and any edge cases that matter. If you want a script that reads a GFF3 file and counts features by type, say that — not just "write a GFF parser."

### Principle 2: Give context about your files and environment
Claude Code can see your files, but it works faster when you point it in the right direction. Mention which file to look at, what format the data is in, or what library you're already using. The more relevant context you give, the less it has to guess.

### Principle 3: Iterate — your first prompt is a draft
Don't expect a perfect solution on the first try. Start with a clear first prompt to get a working skeleton, then follow up with refinements: *"That works, but can you also handle malformed lines without crashing?"* Each follow-up shapes the output closer to what you need.

---

Claude can be wrong. Not often, but it happens — a flag slightly off, a parameter it stated confidently that doesn't match the man page, a biological detail that sounds right but isn't. You'll start to develop a sense for when something needs a second look: when a result surprises you, when a command behaves differently than Claude described, when a claim about the biology doesn't match what you read in the paper. When that feeling comes up, check. It becomes instinct.

## 3. Three Modes of Use

Most of your interactions with Claude Code will fall into one of three patterns:

| Mode | When to use |
|---|---|
| **Write code** | You need a new script, function, or pipeline step and want a working starting point |
| **Explain code** | You encounter code you didn't write (or wrote six months ago) and need to understand what it does |
| **Debug code** | Something broke — an error, unexpected output, or a test failure — and you need to find the cause |

You don't have to declare which mode you're in. Just describe the situation naturally and Claude Code figures out what you need.

## 4. Three Tools You'll Use Every Session

Three custom skills are provided for this training. A **skill** is a Markdown file with structured instructions that Claude follows when you invoke it by name. Skill files live in `.claude/skills/`.

| Invocation | What it does |
|---|---|
| `/log` | Saves a structured session log to `log/session-log.md` — what you worked on, what broke, and what you learned |
| `/debug` | Walks through a 5-step root-cause diagnosis: reproduce → localize → fix → verify → document. Prevents random guessing |
| `/explain` | Explains a concept or tool anchored to the lab's context (organism, file formats, workflows) |

Skills are not magic — they are just Markdown files. You can read them, copy them, and modify them. In Exercise 4 you will do exactly that.
